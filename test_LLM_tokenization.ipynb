```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

engine = create_engine("postgresql://postgres:postgres@localhost:5432/databasename")
connection = engine.connect()


# join some tables, so we have the columns: skill, requested, submit date
query = """SELECT skill, requested, submit_date FROM rfq_consultant_skill 
  LEFT JOIN rfq_consultant_profile ON rfq_consultant_skill.profile_id = rfq_consultant_profile.id 
  LEFT JOIN rfq_request ON rfq_consultant_profile.request_id = rfq_request.id;"""

df = pd.read_sql_query(query, con = engine)
print(df.shape)


column = "skill"
df[column] = df[column].fillna("None")


smalldf = df.head(100)
print(smalldf.shape)
```

```python
categories = ["Taalkennis", "Tester", "Designer", "Team Lead/manager", "analist", "Python", "Javascript", "Java", "C++", "C#", "TypeScript", "Kotlin", "C", "Go", "PHP"]

def f(entry):
    
    result = classifier(entry, categories, multilabel=True)
    print(entry)
    print(result["labels"])
    print(result["scores"])
    print("")
    

smalldf = smalldf.assign(**{"java": smalldf[column].apply(f)})




# roughly: 26s for 10 categories for 100 entries -> 300 000 entries = 21 uur
```
